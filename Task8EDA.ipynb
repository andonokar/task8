# Task 8
Task 8 objective is to analyze data from an EU salary survey for 3 different years: 2018, 2019 and 2020. Exploratory Data Analysis will be used to give meaning to the data between the years

# Loading the dataframes

```python
df2018 = pd.read_csv("EU_Salary/IT Salary Survey EU 2018.csv") 
df2019 = pd.read_csv("EU_Salary/T Salary Survey EU 2019.csv")
df2020 = pd.read_csv("EU_Salary/IT Salary Survey EU  2020.csv")
```

At first look, the data from the 3 dataframe are on different names and one has more information than another. So the first Challenge here is to normalize and reorganize the columns in a way that the first dataframe match second and third information, and then the second and third also match

# Normalizing the columns

## Normalizing 2018 names

```python
df_2018_renamed = df2018.rename(columns={
    "Years of experience": "YearsEXP",
    "Your level": "Seniority",
    "Current Salary": "2018Salary",
    "Salary one year ago": "2017Salary",
    "Salary two years ago": "2016Salary",
    "Are you getting any Stock Options?": "Stocks",
    "Main language at work": "WorkingLanguage",
    "Company size": "CompanySize",
    "Company type": "CompanyType",
})
```

```python
df_2018_renamed.columns
```

## Normalizing 2019 names

```python
df_2019_renamed = df2019.rename(columns={
    "Zeitstempel": "Timestamp",
    "Seniority level": "Seniority",
    "Position (without seniority)": "Position",
    "Years of experience": "YearsEXP",
    "Your main technology / programming language": "ProgrammingLanguage",
    "Yearly brutto salary (without bonus and stocks)": "2019Salary",
    "Yearly bonus": "2019Bonus",
    "Yearly stocks": "2019StocksValue",
    "Yearly brutto salary (without bonus and stocks) one year ago. Only answer if staying in same country": "2018Salary",
    "Yearly bonus one year ago. Only answer if staying in same country": "2018Bonus",
    'Yearly stocks one year ago. Only answer if staying in same country': "2018StocksValue",
    "Number of vacation days": "VacationDays",
    "Number of home office days per month": "HomeOfficeDaysPerMonth",
    "Main language at work": "WorkingLanguage",
    "Company size": "CompanySize",
    "Company type": "CompanyType",
    "Сontract duration": "ContractDuration",
    "Company business sector": "CompanyBusSector",
})
```

```python
df_2019_renamed.columns
```

## Normalizing 2020 names

```python
df_2020_renamed = df2020.rename(columns={
    "Position ": "Position",
    "Total years of experience": "YearsEXP",
    "Years of experience in Germany": "YearsEXPGermany",
    "Seniority level": "Seniority",
    "Your main technology / programming language": "ProgrammingLanguage",
    "Other technologies/programming languages you use often": "SecondaryLanguages",
    "Yearly brutto salary (without bonus and stocks) in EUR": "2020Salary",
    "Yearly bonus + stocks in EUR": "2020BonusAndStocks",
    "Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country": "2019Salary",
    "Annual bonus+stocks one year ago. Only answer if staying in same country": "2020BonusAndStocks",
    "Number of vacation days": "VacationDays",
    "Employment status": "EmploymentStatus",
    "Contract duration": "ContractDuration",
    "Main language at work": "WorkingLanguage",
    "Company size": "CompanySize",
    "Company type": "CompanyType",
    "Have you lost your job due to the coronavirus outbreak?": "CoronavirusLayOff",
    "Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week": "HoursPerWeekCovid",
    "Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR": "MonetarySupport",
})
```

```python
df_2020_renamed.columns
```

#### Salary over the years

#### EXP and Age relation distribution analysis

Considering a lower bound of 20 for age, we can calculate a relationship between year of experience(working) and age. If the result is closer to 1 or bigger, means that for most of the working life they are in IT area

```python
df_2018_renamed["ExpAge"] = df_2018_renamed["YearsEXP"] / (df_2018_renamed["Age"] - 20)
```

```python
sns.displot(data=df_2018_renamed, x="ExpAge", hue="Seniority", kde=True)
```

Here we can see that in the distribution is close to 0.8, which indicates that the majority starts early and tends to keep in IT carrer, the lower ones are people that started the carrer late or had a pause, and higher than 1 are a few outliars that started working in IT before 20 and maintained the carrer.
